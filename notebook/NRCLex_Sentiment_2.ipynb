```python
df = pd.read_csv ('../dataframe/dfc.csv')
long_tweet = ",".join(df["text1"])
```

```python
def word_count(str):
    counts = dict()
    words = str.split()

    for word in words:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1

    return counts
word_count(long_tweet)
```
